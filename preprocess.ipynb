```python
import os
import shutil
data_path = "D:\\wuwen\\ECNU\\大二下\\AI基础\\aimbot\\data\\project-1-at-2024-06-12-02-15-adb11bac"
images_path = os.path.join(data_path,"images")
labels_path = os.path.join(data_path,"labels")
images = os.listdir(images_path)
labels = os.listdir(labels_path)

dst_path = "D:\\wuwen\\ECNU\\大二下\\AI基础\\aimbot\\data"

if os.path.exists(os.path.join(dst_path,"train")):
    shutil.rmtree(os.path.join(dst_path,"train"))
if os.path.exists(os.path.join(dst_path,"val")):
    shutil.rmtree(os.path.join(dst_path,"val"))

lim = int(0.7 * len(images))

for ind, (img, lab) in enumerate(zip(images,labels)):
    print(ind,img,lab)
    tar = "train" if ind < lim else "val"
    tar_path = os.path.join(dst_path,tar)
    if not os.path.exists(tar_path):
        os.mkdir(tar_path)

    tar_images_path = os.path.join(tar_path,"images")
    if not os.path.exists(tar_images_path):
        os.mkdir(tar_images_path)
    src = os.path.join(images_path,img)
    dst = tar_images_path
    shutil.copy(src,dst)

    tar_labels_path = os.path.join(tar_path,"labels")
    if not os.path.exists(tar_labels_path):
        os.mkdir(tar_labels_path)
    src = os.path.join(labels_path,lab)
    dst = tar_labels_path
    shutil.copy(src,dst)
```
